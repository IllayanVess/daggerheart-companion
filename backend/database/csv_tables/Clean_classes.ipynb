````python
import pandas as pd
````

## CLean classes/subclasses/domain

````python
df_classes = pd.read_csv('./srd_classes.csv', sep=';')
df_domain = pd.read_csv('./srd_domain.csv', sep=';')
df_spellcast = pd.read_csv('./srd_spellcast.csv', sep=';')
````

```sql
CREATE TABLE IF NOT EXISTS srd_classes (
    class_id BIGSERIAL PRIMARY KEY,

    class_name TEXT,
    class_description TEXT,
    
    domain1 TEXT ,
    domain2 TEXT ,

    class_item1 TEXT,
    class_item2 TEXT,

    evasion INTEGER,
    starting_hit_points INTEGER,

    hope_feature TEXT,
    class_feature TEXT,

    source_id TEXT REFERENCES srd_sources(id)
);

CREATE TABLE IF NOT EXISTS srd_subclass (
    subclass_id BIGSERIAL PRIMARY KEY,

    class_id BIGINT NOT NULL REFERENCES srd_classes(class_id),

    subclass_name TEXT,
    subclass_description TEXT,

    foundation_feature TEXT,
    specialization_feature TEXT,
    spell_cast_feature TEXT,
    mastery_feature TEXT,

    source_id TEXT REFERENCES srd_sources(id)
);
```

````python
# Create DataFrame from unique class names and domain columns
df_classes_sql = pd.DataFrame({
    'class_name': df_classes['class_name'].unique(),
    'class_description': df_classes['class_name'].unique(),
    'domain1': df_domain['domain1'].values,
    'domain2': df_domain['domain2'].values,
    'class_item_1': df_domain['item1'].values,
    'class_item_2': df_domain['item2'].values,
    'evasion': df_domain['evasion'].values,
    'starting_hit_points': df_domain['starting_hit_points'].values,
    'hope_feature': df_domain['hope_feature'].values,
    'class_feature': df_domain['class_feature'].values,
})
df_classes_sql.to_csv("./df_classes_sql.csv", sep=";", index=False)
````

````python
df_subclass_classes_sql = pd.DataFrame({
    'class_name': pd.concat(
        [df_classes['class_name'], df_classes['class_name']],
        ignore_index=True
    ),
    'subclass' : pd.concat([df_classes['subclass1_name'],
                           df_classes['subclass2_name']],
                           ignore_index=True),
    'subclass_description' : pd.concat([df_classes['subclass1_description'],
                           df_classes['subclass2_description']],
                           ignore_index=True),
    'spellcast_trait_type' : df_spellcast['spellcast_trait'].values,
    'spellcast_trait' : df_spellcast['spellcast_trait_type'].values,
    'foundation_feature' : df_spellcast['foundation_feature'].values,
    'specialization_feature' : df_spellcast['specialization_feature'].values,
    'mastery_feature' : df_spellcast['mastery_feature'].values,

})

df_subclass_classes_sql.to_csv("./df_subclass_classes_sql.csv", sep=";", index=False)
````
